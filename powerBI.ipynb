```python
import psutil
import time
import pyodbc
```

```python
con = pyodbc.connect('Driver={SQL Server};'
                     'Server=LAPTOP-8G2HNKDT\\SQLEXPRESS;'
                     'Database=System_Information;'
                     'Trusted_Connection=yes;')
```

```python
cursor=con.cursor()
```

```python
while 1==1:  #never ending loop
    cpu_usage= psutil.cpu_percent()
    memory_usage=psutil.virtual_memory()[2]

    cpu_interrupts = psutil.cpu_stats() [1]
    cpu_calls = psutil.cpu_stats() [3]
    
    memory_used = psutil.virtual_memory()[3]
    memory_free = psutil.virtual_memory()[4]

    #internet usage information
    bytes_sent = psutil.net_io_counters()[0]
    bytes_received = psutil.net_io_counters()[1]

    #disk usage information
    disk_usage = psutil.disk_usage('/')[3]

    #inserting all information collected in above lines into sql server database
    cursor.execute('insert into Performance values (GETDATE(),'
                   +str(cpu_usage)+','
                   +str(memory_usage)+','
                   +str(cpu_interrupts)+','
                   +str(cpu_calls)+','
                   +str(memory_used)+','
                   +str(memory_free)+','
                   +str(bytes_sent)+','
                   +str(bytes_received)+','
                   +str(disk_usage)+')'
                   )
    
    con.commit()  #to finalise insertion and make the sql server database accessible to other application
    
    print(memory_usage)
    time.sleep(1)
```
